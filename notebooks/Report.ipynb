```python
df.doxc_notspace.str.repeat('.*\/data\/([0-9]{4})\/.*\/((?:IntheNews)(\d+).(\d*))\.docx$')
```

```python
df['doc_year'] = df.doxc_notspace.str.extract(r'(?:\/data\/)([0-9]{4})')
```

```python
df['doc_date'] = df.doxc_notspace.str.extract(r'-[a-zA-Z]*(\d*)|-\d* ')
```

```python
df['doc_month'] = df.doxc_notspace.str.extract(r'\/\d+\/\d?([a-zA-Z]*)\/')
```
