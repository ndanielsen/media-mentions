```python
import pandas as pd
import tldextract
```

```python
df = pd.read_pickle('irexmedia.pickle')
```

```python
# df.info()
```

```python
# (ext.subdomain, ext.domain, ext.suffix)
df['domain_info'] = df.Url.apply(tldextract.extract)
cols = ['subdomain', 'domain', 'domain_suffix']
df['subdomain'] = df.domain_info.str[0]
df['domain'] =  df.domain_info.str[1]
df['domain_suffix'] =  df.domain_info.str[2]
```

```python
import bs4

def html_lang(n):
    try:
        soup = bs4.BeautifulSoup(n)
        lang = soup.find('html')['lang']
    except:
        lang = 'Unknown'
    return lang
```

```python
df['html_lang'] = df.html_raw.apply(html_lang)
```

```python
urls = df.groupby('domain').Url.count().reset_index(0)
```

```python
urls.sort_values('Url', ascending=False)
```

```python
us_suffix = ['gov', 'edu']

us_suffix_list = df.domain_suffix.apply(lambda x: x == us_suffix[0] or x == us_suffix[1])
```
