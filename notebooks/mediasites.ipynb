```python
import json
from pprint import pprint

with open(datafile, 'rU') as f:
    data_list = list(f)
```

```python
domains = {json.loads(data)['url']: json.loads(data) for data in data_list}
```

```python
df = pd.DataFrame(domains)
df = df.T.reset_index()
```

```python
try:
    del df['index']
    del df['priority']
except:
    pass
df.describe()
```

```python
df
```

```python
df['domain_info'] = df.url.apply(tldextract.extract)
df['subdomain'] = df.domain_info.str[0]
df['domain'] =  df.domain_info.str[1]
df['domain_suffix'] =  df.domain_info.str[2]

```

```python
df['num'] = 1
domains = df.groupby('domain').count()
```
